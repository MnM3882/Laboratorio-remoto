```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.fftpack
import scipy.optimize
```

```python
distances = [83, 100, 116, 133, 149, 166, 183, 199, 216, 233, 249, 265, 281, 298, 314, 331, 346]
d = np.array(distances)/10.0
V = np.array([])
Vstd = np.array([])
Vstd50 = np.array([])
P = np.array([0, 39, 78, 117, 156, 195, 234, 273, 312, 351, 390, 429, 468, 507, 546, 585, 624])

fx = np.linspace(0.0, 1.0/(2.0*4.999/5000), 5000/2)
fy = np.zeros(5000)

for i in range(len(distances)):
    df=pd.read_csv(str(distances[i])+'mm/anag2.csv', header=None, squeeze = True)*5/4095
    V = np.append(V,np.mean(df[1:]))
    Vstd = np.append(Vstd,np.mean(df[1:]))
    
    V_50ms=[np.mean(df[1:50])]
    
    for j in range(50):
        V_50ms.append(np.mean(df[50*(j+1):50*(j+2)]))
    
    Vstd50 = np.append(Vstd50,np.std(V_50ms))
    
    
    df.plot.kde(figsize=(16,8), xlim=(500*5/4095, 2500*5/4095))
    fy = fy + scipy.fftpack.fft(df.values)
```

```python
dfit = np.linspace(d[0], d[-1], 50)

def rsi_fun(x, a, b):
    return a * x**-2 + b

popt_Vr, pconv_Vr = scipy.optimize.curve_fit(rsi_fun, d, V)
ss_tot_Vr = np.sum((V-np.mean(V))**2)
residuals_Vr = V- rsi_fun(d, *popt_Vr)
ss_res_Vr = np.sum(residuals_Vr**2)
r_squared_Vr = 1 - (ss_res_Vr / ss_tot_Vr)

plt.plot(dfit, rsi_fun(dfit, *popt_Vr), 'g--', 
         label='Fit: a*x^-2 + b \n a=%5.3f, b=%5.3f'%tuple(popt_Vr)+'\n r^2=%5.3f'%r_squared_Vr)
plt.scatter(d,V)
plt.legend()
plt.show()

print(r_squared_Vr)
```

```python
def exp_fun(x, a, b, c):
    return a * np.exp(-b*x) + c

popt_Ve, pconv_Ve = scipy.optimize.curve_fit(exp_fun, d, V)
ss_tot_Ve = np.sum((V-np.mean(V))**2)
residuals_Ve = V- exp_fun(d, *popt_Ve)
ss_res_Ve = np.sum(residuals_Ve**2)
r_squared_Ve = 1 - (ss_res_Ve / ss_tot_Ve)
```

```python
fig, ax = plt.subplots(1, 1, sharex=False, figsize=(12,8))

ax.errorbar(x=d*10, xerr=1, y=V, yerr=Vstd50, ecolor='red', capsize=3 , marker='.', mew=2, ls='None')

ax.plot(dfit*10, exp_fun(dfit, *popt_Ve), 'g--', 
         label='Fit: a*e^-b*x/10 + c \n a=%5.3f, b=%5.3f, c=%5.3f'%tuple(popt_Ve)+'\n r^2=%5.4f'%r_squared_Ve)

ax.legend()

ax.set_xlabel('Distancia [mm]')
ax.set_ylabel('Voltaje [V]')

fig.tight_layout()
```
